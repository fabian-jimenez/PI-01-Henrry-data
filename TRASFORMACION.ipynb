```python
#Importar las librerias 
import pandas as pd 
import numpy as np
import re 
```

```python
# creaccion de dataframes apartir de los archibos csv 
dfmovies= pd.read_csv('C:/Users/FABIAN/Desktop/PI01FINAL/movies_dataset.csv')
dfcredits = pd.read_csv('C:/Users/FABIAN/Desktop/PI01FINAL/credits.csv')
#dfmovies_dataset =pd.read_excel('C:/Users/FABIAN/Desktop/PI01FINAL/movies_dataset.xlsx')
dfmovies.head ()
```

```python
# eliminar columnas que no se utilizaran video,imdb_id,adult,original_title,poster_path y homepage 
dfmovies = dfmovies.drop(['video','imdb_id','adult','original_title','homepage'], axis=1)

dfmovies.head()
```

```python
#cambiar valores nulos de las columnas revenue, budget

dfmovies['budget']= dfmovies['budget'].fillna('0')
dfmovies['revenue']= dfmovies['revenue'].fillna('0')
dfcredits.head()
```

```python
#eliminar datos nulos en la columna release date 

dfmovies['release_date'].dropna(inplace=True)
```

```python
# formato a fecha  de la columna release_date 
dfmovies['release_date']= pd.to_datetime(dfmovies['release_date'],errors='coerce') 
#creacion de la clumna release_year donde solo muestra el año de la pelicula 
dfmovies['release_year']= dfmovies['release_date'].dt.year

dfmovies.head()
```

```python
#crezcion de la cliumna retur que debuelve la operacion entre revenue / budget

# se crea la columna con valores 0 
dfmovies['return']= 0 
#convertir los datos de las columas en valores numerios  
dfmovies['budget']=pd.to_numeric(dfmovies['budget'],errors='coerce')
dfmovies['revenue']=pd.to_numeric(dfmovies['revenue'],errors='coerce')
#calclula la divicion solo cuando la divicion es posible 
#crear una mascar para poder hacer la combinacion de ambas condiciones 
mask= (dfmovies['revenue'].notnull()) & (dfmovies['budget'].notnull())

dfmovies.loc[mask, 'return'] = dfmovies.loc[mask,'revenue'] / dfmovies.loc[mask,'budget']

dfmovies.head()
```

```python
# funcion de desanidar los datos de cada in a de las columnas 

def extraer_datos(df, columna):
    # Función para extraer los valores de una cadena, lista o diccionario
    def extraer_valores(valor):
        if isinstance(valor, str):
            # Si es una cadena, extraer los valores con la expresión regular
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, valor)
            return coincidencias
        elif isinstance(valor, list):
            # Si es una lista, extraer los valores de cada elemento de la lista
            valores = []
            for elemento in valor:
                if 'name' in elemento:
                    valores.append(elemento['name'])
            return valores
        elif isinstance(valor, dict):
            # Si es un diccionario, extraer los valores de la clave 'name'
            if 'name' in valor:
                return valor['name']
        return None
    
    # Aplicar la función a la columna y sobrescribir la columna
    df[columna] = df[columna].apply(extraer_valores)
    
    return df
# aplicacion de la funcion a las columnas anidadas 

dfmovies = extraer_datos(dfmovies, 'belongs_to_collection')
dfmovies = extraer_datos(dfmovies, 'genres')
dfmovies = extraer_datos(dfmovies,'production_companies')
dfmovies = extraer_datos(dfmovies, 'production_countries')
dfmovies = extraer_datos(dfmovies, 'spoken_languages')


dfmovies.head()
```

```python
# aplicacion de la funcion a el otro df 

dfcredits = extraer_datos(dfcredits, 'cast')
dfcredits = extraer_datos(dfcredits, 'crew')

dfcredits.head()
```

```python
#igualar los valor de id en letras (str) para poderlos unir 
dfcredits['id']=dfcredits['id'].astype(str)
dfmovies['id'] = dfmovies['id'].astype(str)
# unir los csv 
dfmovies_final = pd.merge (dfmovies,dfcredits, on='id', how= 'left')
dfmovies_final.head()
```

```python
# creacion del csv ya con el etl listo para hacer la api apertir de ese  csv y el eda 

dfmovies_final.to_csv ('C:/Users/FABIAN/Desktop/PI01FINAL/Movies_final_ETL.csv',index=False)
```
